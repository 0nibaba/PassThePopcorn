```python
import pandas as pd
import numpy as np
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
credits = credits.drop('title', axis=1)
movies.rename(columns = {'id':'movie_id'},inplace=True)
movies = movies.merge(credits,on='movie_id')
movies = movies[['movie_id', 'title', 'genres', 'keywords', 'overview', 'cast', 'crew']]
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
import ast
def tags(obj):
  A = []
  for i in ast.literal_eval(obj):
    A.append(i['name'])
  return A
```

```python
movies['genres'] = movies['genres'].apply(tags)
```

```python
movies['keywords'] = movies['keywords'].apply(tags)
```

```python
def director(obj):
  A = []
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      A.append(i['name'])
      break;
  return A;
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
def actors (obj):
    count = 0;
    A = []
    for i in ast.literal_eval(obj):
      if(count < 3):
        A.append(i['name'])
        count += 1
      else:
        break
    return A
```

```python
movies['cast'] = movies['cast'].apply(actors)
```

```python
movies.rename(columns = {'crew':'director'},inplace=True)
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['director'] = movies['director'].apply(lambda x:[i.replace(' ','') for i in x])
movies
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['director'] + movies['overview']
movies = movies[['movie_id', 'title', 'tags']]
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
movies['tags']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
movies
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem (text):
  A = []
  for i in text.split():
    A.append(ps.stem(i))
  return " ".join(A)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7500,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarityScores = cosine_similarity(vectors)
```

```python
def recommend (movie):
  movieIndex = movies[movies['title'] == movie].index[0]
  vectorDistances = similarityScores[movieIndex]
  similarMovies = sorted(list(enumerate(vectorDistances)), reverse=True, key=lambda x:x[1])[1:6]

  for i in similarMovies:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
